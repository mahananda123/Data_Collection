###Collect shopping data from sqlite
The attached shopping.sqlite file contains a dummy shopping dataset.

###Establish connection with data base

```python
import sqlite3
```

```python
connection=sqlite3.connect('/content/shopping.sqlite')
cursor=connection.cursor()
```

```python
query='''
select name from sqlite_master
where type='table';
'''
```

```python
cursor.execute(query)
results=cursor.fetchall()
results
```

```python
#retrive information from data base
query='''
select * from customer_shopping_data
limit 3;
'''
cursor.execute(query)
result=cursor.fetchall()
result
```

Fetch all the records

```python
query='''
select * from customer_shopping_data;
'''
cursor.execute(query)
resultx=cursor.fetchall()
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(resultx)
df.info()
```
