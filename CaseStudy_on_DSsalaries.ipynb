###Concatenation

documenation:

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.concat.html


```python
df1=df[['company_location','job_title','salary_in_usd']].sample(n=5)
df2=df[['company_location','job_title','salary_in_usd']].sample(n=5)
df3=df[['company_location','job_title','salary_in_usd']].sample(n=5)
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
df_cat1=pd.concat([df1,df2,df3],axis=0)
df_cat1
```

```python
df_cat1.info()
```

```python
df_cat2=pd.concat([df1,df2,df3],axis=1)
df_cat2
```

```python
df_cat2.info()
```

###Merging
Documentation:
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html



```python
df1=df[['company_location','experience_level','salary_in_usd']][0:5]
df1
```

```python
df2=df[['company_location','job_title','salary_in_usd']][:5]
df2
```

```python
pd.merge(df1,df2,on='company_location',how='inner')
```

```python
pd.merge(df1,df2,on='company_location',how='inner').drop_duplicates()
```

```python
df3=df[['company_location','job_title','experience_level']][2:6]
df3
```

```python
pd.merge(df1,df3,on='company_location',how='inner').drop_duplicates()
```

```python
pd.merge(df1,df3,on='company_location',how='outer').drop_duplicates()
```

###Joining
documentation:
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.join.htm

```python
df1=df[['experience_level']][:5]
df1
```

```python
df2=df[['job_title']][2:7]
df2
```

```python
df1.join(df2,how='left').drop_duplicates()
```

```python
df1.join(df2,how='right').drop_duplicates()
```

```python
df1.join(df2,how='inner').drop_duplicates()
```

```python
df1.join(df2,how='outer').drop_duplicates()
```
