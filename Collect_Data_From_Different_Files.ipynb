```python
# import pandas library
import pandas as pd
```

###Read CSV files using read_csv function


```python
df=pd.read_csv('/content/ds_salaries.csv')
df.head()
```

```python
df.info()
```

###We can manipulate arguments for specific csv file

```python
df=pd.read_csv('/content/ds_salaries.csv',header=None,skiprows=1,skipfooter=3)
df.head()
```

```python
df.info()
```

###The txt follows csv format, then it can be read as a csv file

```python
df=pd.read_csv('/content/ds_salaries.txt')
df
```

###Excel

```python
df=pd.read_excel('/content/ds_salaries.xlsx')
df
```

```python
df.info()
```

###json

```python
df=pd.read_json('/content/ds_salaries.json')
df
```

```python
df.info()
```

###XML

```python
df=pd.read_xml('/content/ds_salaries.xml')
df.head()
```

```python
df.info()
```
